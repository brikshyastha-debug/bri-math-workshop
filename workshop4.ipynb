```python
#task1
import matplotlib.pyplot as plt
import random
x=[random.randint(0,50) for _ in range(10)]
y=[random.randint(0,50) for _ in range(10)]
colors=["red","green","blue","yellow","pink","orange","purple","brown","cyan","magenta"]

plt.scatter(x,y,c=colors)

plt.xlabel("time taken in seconds")
plt.ylabel("distance reached")
plt.title("buzzer game results")

plt.show()
```

```python
import matplotlib.pyplot as plt

animals = ["bear","dolphin","whale"]
a=[8,150,80]
b=[54,77,54]
c=[93,32,100]
d=[116,11,76]
e=[137,11,93]
f=[184,6,72]

plt.figure(figsize=(16,20))


plt.subplot(1,3,1)
plt.pie(a, labels=animals)
plt.title("2017")

plt.subplot(1,3,2)
plt.pie(b, labels=animals)
plt.title("2018")

plt.subplot(1,3,3)
plt.pie(c, labels=animals)
plt.title("2019")

plt.subplot(2,3,4)
plt.pie(d, labels=animals)
plt.title("2020")

plt.subplot(2,3,5)
plt.pie(e, labels=animals)
plt.title("2021")

plt.subplot(2,3,6)
plt.pie(f, labels=animals)
plt.title("2022")

plt.show()
```

```python
import matplotlib.pyplot as plt

cars = ["Red","Silver","Blue","White","Green"]
values = [123,230,78,193,12]

#a
plt.bar(cars, values)
plt.title("Bar Chart")
plt.show()

#b
plt.bar(cars, values, width=0.3)
plt.title("Width = 0.3")
plt.show()

#c
plt.barh(cars, values)
plt.title("Horizontal Bar")
plt.show()

#d
colors = ["red","pink","blue","yellow","green"]
plt.bar(cars, values, color=colors)
plt.title("Colored Bars")
plt.show()

#e
plt.pie(values, labels=cars)
plt.title("Pie Chart")
plt.show()

#f
myexplode = [0.3,0,0,0,0]
mycolours = ["magenta","green","cyan","red","yellow"]
plt.figure()
plt.pie(values, labels=cars, explode=myexplode, shadow=True, colors=mycolours)
plt.title("Pie (explode + shadow + colors)")
plt.legend(title="Car Colours:")
plt.show()
```
