```python
import numpy as np
import matplotlib.pyplot as plt

origin = np.array([0,0])
V = np.array([1,1])
W = np.array([-2.0,-1])
fig, ax = plt.subplots()

ax.quiver(origin[0], origin[1], V[0], V[1], angles='xy', scale_units='xy', scale=1, color='r')
ax.quiver(origin[0], origin[1], W[0], W[1], angles='xy', scale_units='xy', scale=1, color='b')
ax.text(origin[0]-0.1, origin[1]+0.1, "O")
ax.text(V[0], V[1], "v")
ax.text(W[0]-0.125, W[1]-0.1, "w")
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.grid()
plt.show()
```

```python
#task6 a)
import numpy as np
import matplotlib.pyplot as plt

u = np.array([1, 0.25, 0])
v = np.array([0, 1, 0.25])
w = np.array([0.25, 0, 1])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

origin = [0,0,0]

ax.quiver(*origin, *u)
ax.quiver(*origin, *v)
ax.quiver(*origin, *w)

plt.show()
```

```python
#b)
import matplotlib.pyplot as plt
import numpy as np

U = np.array([-0.5, 0.5])
V = np.array([1, 1])
W = U + V
origin = np.array([0, 0])

plt.figure()
plt.quiver(*origin, *U, angles='xy', scale_units='xy', scale=1)
plt.quiver(*origin, *V, angles='xy', scale_units='xy', scale=1)
plt.quiver(*origin, *W, angles='xy', scale_units='xy', scale=1)

plt.xlim(-1, 2)
plt.ylim(-1, 2)

plt.axhline(0)
plt.axvline(0)
plt.grid()

plt.text(U[0], U[1], 'U')
plt.text(V[0], V[1], 'V')
plt.text(W[0], W[1], 'W')

plt.show()
```

```python
#c
import numpy as np

def magnitude(v):
    return np.sqrt(np.sum(v**2))

U = np.array([1.5,1.5,2.0])
print(magnitude(U))
```

```python
#d
def direction_cosines(v):
    mag = magnitude(v)
    return v / mag

O = np.array([1.5,1.5,2.0])
print(direction_cosines(O))
```

```python
#e
def dot_product(a,b):
    return np.dot(a,b)

a = np.array([1.5,1.5,2.0])
b = np.array([-1.5,0.5,2.0])

print(dot_product(a,b))
```

```python
#f
def cross_product(a,b):
    return np.cross(a,b)

a = np.array([1.5,1.5,2.0])
b = np.array([-1.5,0.5,2.0])

print(cross_product(a,b))
```
