Write a program to prompt the user for hours and rate per hour using input( ) to compute
gross pay. Pay the hourly rate for the hours up to 40 and 1.5 times the hourly rate for all
hours worked above 40 hours. Use 45 hours and a rate of 10.50 per hour to test the program
(the pay should be 498.75)

```python
hrs = input("enter hours: ")
h = float(hrs)
rate = input("enter the rate per hour: ")
r = float(rate)

if h <= 40:
    pay = h * r
elif h > 40:
    pay = (40 * r) + ((h - 40) * r * 1.5)

print("the pay is " + str(pay))
```

```python
def fib(n):
  a,b=0,1
  series=[]
  for i in range (n):
    series.append(a)
    a,b=b,a+b
  return series
print (fib(10))
```

```python
def mypow(x, n):
    i = 0
    result = 1
    while i < n:
        result = result * x
        i = i + 1
    return result
print(mypow(2, 3))
```

```python
def SumDigits(n):
    total = 0
    while n > 0:
        digit = n % 10
        total = total + digit
        n = n // 10
    print(total)
SumDigits(123)
```

```python
def squaredSum(arr):
    total = 0
    i = 0

    while i < len(arr):
        total = total + arr[i] * arr[i]
        i = i + 1

    return total
print(squaredSum([2,3,4]))
```
