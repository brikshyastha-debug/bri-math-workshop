```python
def unique_list (lst):
  unique=[]
  for item in lst:
    if item not in unique:
        unique.append(item)
  return unique


sample= [1,2,3,3,3,3,4,5]
print(unique_list(sample))
```

```python
def prime (num):
  if num<=1:
    return "not prime"
  for i in range (2,num):
    if num%i==0:
      return "not prime"

  return "prime"

num= int(input("enter a number: "))
print (prime(num))
```

```python
def sumofDivision(num):
  total=0
  for i in range (1,num):
    if num %i==0:
      total +=i

  if total ==num:
    return "perfect number"
  else:
    return "not perfect number"

num= int(input("enter a number: "))
print(sumofDivision(num))


```

```python
def decimal_to_binary(num):
  if num==0:
    return "0"
  binary=""
  while num>0:
    remainder= num%2
    binary =str(remainder)+binary
    num=num//2
  return binary
n=int(input("enter a number: "))
print(decimal_to_binary(n))
```

```python
from string import digits
def prodigits(num):
  product=1
  if num<0:
    num=abs(num)
  if num==0:
    return 0
  while num>0:
    digit= num%10
    product*= digit
    num//= 10
  return product
number=98
print(f"the product digits of {number} is: {prodigits(number)}")
```

```python
def multiplyAll(*args):
    result = 1
    for num in args:
        result *= num
    return result

print(multiplyAll(1, 2, 3, 4, 5))
print(multiplyAll(12, 13, 14))
```
