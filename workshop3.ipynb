```python
import matplotlib.pyplot as plt
import math
x = []
y = []
start = -4 * math.pi
end = 4 * math.pi
step = 0.1
i = start
while i <= end:
    x.append(i)
    y.append(math.sin(i))
    i += step
plt.plot(x, y, color="red", label="sin function")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("y = sin(x), -4π ≤ x ≤ 4π")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import math
x = []
y = []
r = 4
center = [2, 2]
theta_start = 0.0
theta_end = 2 * math.pi
step_size = 0.01
theta = theta_start
while theta <= theta_end:
    x.append(center[0] + r * math.cos(theta))
    y.append(center[1] + r * math.sin(theta))
    theta += step_size
plt.figure()
plt.plot(center[0], center[1], '.')
plt.plot(x, y, '-')
plt.gca().set_aspect('equal', adjustable='box')
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
#exercise6
import matplotlib.pyplot as plt
import math
x = []
y1 = []
y2 = []
y3 = []
for i in range(20):
    x.append(i)
for value in x:
    y1.append(5*value**3 + 2*value - 1)
    y2.append(-2*value**3 + value**2 + 100)
    y3.append(2*math.pi*value + 20)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Plot of Three Functions")
plt.legend()
plt.grid()
plt.show()
```

```python
#exercise 7
import matplotlib.pyplot as plt
import math
x = []
y1 = []
y2 = []
y3 = []
for i in range(20):
    x.append(i)
for val in x:
    y1.append(5*val**3 + 2*val - 1)
    y2.append(-2*val**3 + val**2 + 100)
    y3.append(2*math.pi*val + 20)
#create subplots (3 rows, 1 column)
plt.subplot(3, 1, 1)
plt.plot(x, y1)
plt.title("f(x) = 5x^3 + 2x - 1")
plt.grid()
plt.subplot(3, 1, 2)
plt.plot(x, y2)
plt.title("f(x) = -2x^3 + x^2 + 100")
plt.grid()
plt.subplot(3, 1, 3)
plt.plot(x, y3)
plt.title("f(x) = 2πx + 20")
plt.grid()
plt.tight_layout()  # avoids overlap
plt.show()
```

```python
#exercise9
import matplotlib.pyplot as plt
import math
x = []
y = []
start = -3 * math.pi
end = 3 * math.pi
step = 0.1
i = start
while i <= end:
    x.append(i)
    y.append(math.cos(i))
    i += step
plt.plot(x, y, label="cos(x)", color="blue")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("y = cos(x), -3π ≤ x ≤ 3π")
plt.grid()
plt.legend()
plt.show()
```

```python
#exercise10
import matplotlib.pyplot as plt
import math
x = []
y = []
a = 2
b = 2
r = 4
theta = 0
end = 2 * math.pi
step = 0.01
while theta <= end:
    x.append(a + r * math.cos(theta))
    y.append(b + r * math.sin(theta))
    theta += step
plt.plot(x, y)
plt.plot(a, b,'ro')
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Circle with center (2,2) and radius 4")
plt.grid()
plt.legend()
plt.show()
```

```python
#exercise11
import matplotlib.pyplot as plt
import math
xc = []
yc = []
xe = []
ye = []
theta = 0
end = 2 * math.pi
step = 0.01
while theta <= end:
    xc.append(2 + 4 * math.cos(theta))
    yc.append(2 + 4 * math.sin(theta))
    xe.append(2 + 6 * math.cos(theta))
    ye.append(2 + 3 * math.sin(theta))
    theta += step
plt.plot(xc, yc, label="Circle", color="blue")
plt.plot(xe, ye, label="Ellipse", color="red")
plt.plot(2, 2, 'ko', label="Center")
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Circle and Ellipse on Same Axes")
plt.grid()
plt.legend()
plt.show()
```

```python
#exercise12
import numpy as np
import matplotlib.pyplot as plt
p = 2
q = 1
a = 0.5
x = np.linspace(-5, 7, 400)
y = p + 4*a*(x - q)**2
plt.plot(x, y, label="Parabola")
plt.scatter(q, p, color='red', label="Vertex (q, p)")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Parabola with Vertex (p, q)")
plt.legend()
plt.grid()
plt.show()
```
